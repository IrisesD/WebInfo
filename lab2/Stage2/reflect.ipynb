```python
id2entities = {}
with open("douban2fb.txt", 'r') as f:
    while True:
        line = f.readline()
        if not line:
            break
        else:
            line_list = line.split()
            id2entities[line_list[0]] = line_list[1]
for key in id2entities.keys():
    print(f"{key}:{id2entities[key]}")
```

```python
id2index = {}
with open("movie_id_map.txt", 'r') as f:
    while True:
        line = f.readline()
        if not line:
            break
        else:
            line_list = line.split()
            id2index[line_list[0]] = line_list[1]
for key in id2index.keys():
    print(f"{key}:{id2index[key]}")
```

```python
entities2index = {}
for key in id2entities:
    entities2index[id2entities[key]] = id2index[key]
for key in entities2index:
    print(f"{key}:{entities2index[key]}")
```

```python
import json
with open("graph_2step.json", 'r') as f:
    graph_2step = json.load(f)
```

```python
print(len(graph_2step))
```

```python
relation2index = {}
relindex = 0
entityindex = len(entities2index)
print(entityindex)
triples = []
for entity in graph_2step.keys():
    if entity not in entities2index.keys():
        entities2index[entity] = str(entityindex)
        entityindex += 1
    first_hop = graph_2step[entity]
    for rel in first_hop.keys():
        if rel not in relation2index.keys():
            relation2index[rel] = str(relindex)
            relindex += 1
        for first_entity in first_hop[rel].keys():
            if first_entity not in entities2index.keys():
                entities2index[first_entity] = str(entityindex)
                entityindex += 1
            triples.append([entities2index[entity], relation2index[rel], entities2index[first_entity]])
            # process the first hop

            second_hop = first_hop[rel][first_entity]
            for rel2 in second_hop.keys():
                if rel2 not in relation2index.keys():
                    relation2index[rel2] = str(relindex)
                    relindex += 1
                for second_entity in second_hop[rel2].keys():
                    if second_entity not in entities2index.keys():
                        entities2index[second_entity] = str(entityindex)
                        entityindex += 1
                    triples.append([entities2index[first_entity], relation2index[rel2], entities2index[second_entity]])
```
